```python
import os 
import pandas as pd
import pyarrow as pa 
#Leemos el dataset
df = pd.read_csv(r"C:\Users\WalterECalcagnoLucar\OneDrive - inegocios.cl\Unifrutti_PowerBIAv\CSV\Online Sales.csv", header="infer" , sep=';')

display(df)
display(df.dtypes)
```

```python
#hacemos los cambios de tipos de datos que necesitamos antes de parquetizar
df['Date of Sale'] = pd.to_datetime(df['Date of Sale'], infer_datetime_format=True)
df['Unit Cost'] = df['Unit Cost'].replace(',','.', regex=True)
df['Unit Cost'] = pd.to_numeric(df['Unit Cost'])
df['Unit Price'] = df['Unit Price'].replace(',','.', regex=True)
df['Unit Price'] = pd.to_numeric(df['Unit Price'])
df['Discount Amount']=df['Discount Amount'].replace(',','.', regex=True)
df['Discount Amount']=pd.to_numeric(df['Discount Amount'])

display(df.dtypes)
display(df)
```

```python
# parquetizaremos con una particion por año
##primero creamos la columna año en el dataset para usarlo como guia en la particion
df['year'] = pd.to_datetime(df['Date of Sale']).dt.year

#luego especificamos el directorio donde guardaremos los archivos parquetizados
destino = r"C:\Users\WalterECalcagnoLucar\OneDrive - inegocios.cl\Unifrutti_PowerBIAv\PARQUET"
parquet_file = os.path.join(destino,'dataset.snappy.parquet')

#escribimos el parquet en la ruta definida anteriormente y lo particionamos por la columna year
df.to_parquet(parquet_file, engine='pyarrow', compression='snappy', partition_cols=['year'], index=False  )

display(df)
```
